```python
!pip install -q -U google-generativeai

import google.generativeai as genai
from google.colab import userdata

# Importa a biblioteca para usar a API do Google Generative AI

# Define a chave da API
api_key = userdata.get("secret_key")
genai.configure(api_key=api_key)

# Configura as configurações de geração de texto
generation_config = {
    "temperature": 0.5,  # Controla a criatividade do modelo
    "top_p": 0.95,  # Limita a probabilidade das palavras escolhidas
    "top_k": 0,  # Não limita o número de palavras consideradas
    "max_output_tokens": 8192,  # Define o tamanho máximo da resposta
}

# Define as configurações de segurança para o modelo
safety_settings = [
    {
        "category": "HARM_CATEGORY_HARASSMENT",
        "threshold": "BLOCK_MEDIUM_AND_ABOVE"  # Bloqueia conteúdo ofensivo
    },
    {
        "category": "HARM_CATEGORY_HATE_SPEECH",
        "threshold": "BLOCK_MEDIUM_AND_ABOVE"  # Bloqueia discurso de ódio
    },
    {
        "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
        "threshold": "BLOCK_MEDIUM_AND_ABOVE"  # Bloqueia conteúdo sexualmente explícito
    },
    {
        "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
        "threshold": "BLOCK_MEDIUM_AND_ABOVE"  # Bloqueia conteúdo perigoso
    },
]

# Inicializa o modelo de linguagem
model = genai.GenerativeModel(
    model_name="gemini-1.5-pro-latest",  # Nome do modelo a ser utilizado
    generation_config=generation_config,  # Aplica as configurações de geração
    safety_settings=safety_settings  # Aplica as configurações de segurança
)


# Mensagem de boas-vindas ao usuário
print("Seja bem vindo! Eu sou seu Assistente de Estudos. Qual a disciplina que vamos estudar agora?")
print()

# Lista as opções de disciplinas disponíveis
print("Artes")
print("Biologia")
print("Espanhol")
print("Filosofia")
print("Física")
print("Francês")
print("Geografia")
print("História")
print("Inglês")
print("Matemática")
print("Português")
print("Química")
print()

# Recebe a disciplina escolhida pelo usuário
DISCIPLINA = input("Digite a disciplina: ")
print("------------------------------------------------------------------------")

# Solicita o conteúdo específico da disciplina
print("Qual o conteúdo de estudo?")
CONTEUDO = input()
print("------------------------------------------------------------------------")

# Solicita a dúvida do usuário
print("Qual a sua dúvida?")
DUVIDA = input()
print("------------------------------------------------------------------------")

# Cria um prompt para o modelo de linguagem
prompt = f"Crie um texto de {DISCIPLINA} sobre {CONTEUDO} que esclareça {DUVIDA}"

# Gera uma resposta para a dúvida do usuário
response = model.generate_content(prompt)
print(response.text)
print("------------------------------------------------------------------------")

# Pergunta se a resposta esclareceu a dúvida do usuário
ESCLARECIMENTO = input("Sua dúvida foi esclarecida? (Sim, Não, Aprofundar): ")
print()

# Loop para oferecer opções caso a dúvida não tenha sido esclarecida
while ESCLARECIMENTO != "Sim":
    if ESCLARECIMENTO == "Não":
        # Gera uma resposta mais simples
        prompt = f"Crie um texto mais simples, com menos informações, que esclareça {DUVIDA}"
        response = model.generate_content(prompt)
        print("Texto Mais Simples!","\n")
        print(response.text)
        print("------------------------------------------------------------------------")
    elif ESCLARECIMENTO == "Aprofundar":
        # Gera uma resposta mais aprofundada
        prompt = f"Crie um texto com mais informações e detalhes que complementem {DUVIDA}"
        response = model.generate_content(prompt)
        print("Mais Informações e Detalhes!","\n")
        print(response.text)
        print("------------------------------------------------------------------------")
    # Pergunta novamente se a dúvida foi esclarecida
    ESCLARECIMENTO = input("Sua dúvida foi esclarecida? (Sim, Não, Aprofundar): ")
    print()

# Mensagem de encerramento
print("Fico feliz em ter ajudado! Bons estudos!")
```
